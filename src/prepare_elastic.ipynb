```python
NUMBER_OF_WORDS = 100
PATH_KNOWLEDGE_BASE = 'news_envBR_bm25.csv'
INDEX_KNOWLEDGE_BASE = "news_ambiente"
#INDEX_KNOWLEDGE_BASE = "wiki_clean_ambiente"
#INDEX_KNOWLEDGE_BASE = "wiki_news_ambiente"
```

```python
dicts = []
with open(PATH_KNOWLEDGE_BASE, 'r', encoding='utf-8') as file:
    for line in file:
        content = line.split('\t')
        title = content[0]
        text = content[1]
        dicts.append({'text' : text, 'meta' : {'title' : title}})

from haystack.preprocessor import PreProcessor
processor = PreProcessor(
    clean_empty_lines=True,
    clean_whitespace=True,
    clean_header_footer=True,
    split_by="word",
    split_length=NUMBER_OF_WORDS,
    split_respect_sentence_boundary=False,
    split_overlap=0
)
docs = []
cont=0

for dict1 in dicts:
    if cont %100 ==0:
        print(cont)
    cont+=1
    doc = processor.process(dict1)
    docs = docs+doc
```

```python
from haystack.document_store.elasticsearch import ElasticsearchDocumentStore
document_store = ElasticsearchDocumentStore(host="localhost", username="", password="", index=INDEX_KNOWLEDGE_BASE)
document_store.write_documents(docs,batch_size=1000)
```
