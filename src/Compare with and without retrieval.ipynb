```python
import pandas as pd
```

```python
df_retrieval = pd.read_csv("TEST_RESULTS_QA-ptt5-base-wiki-k10_checkpoint-9000.csv")
```

```python
df_without_retrieval = pd.read_csv("TEST_RESULTS_QA-ptt5-base-without-retrieval_checkpoint-9000.csv")
```

```python
df_retrieval.head()
```

```python
df_retrieval = df_retrieval.rename(columns={'gen_answer': 'gen_answer_news', 'QA-ptt5-base-wiki-k10/checkpoint-9000_f1':'f1_wiki', 'QA-ptt5-base-wiki-k10/checkpoint-9000_rouge_l_f':'rougeL_wiki','QA-ptt5-base-wiki-k10/checkpoint-9000_em':'em_wiki'})
```

```python
df_without_retrieval.head()
```

```python
df_without_retrieval = df_without_retrieval.rename(columns={'gen_answer': 'gen_answer_without_retrieval', 'QA-ptt5-base-without-retrieval/checkpoint-9000_f1':'f1_without_retrieval', 'QA-ptt5-base-without-retrieval/checkpoint-9000_rouge_l_f':'rougeL_without_retrieval','QA-ptt5-base-without-retrieval/checkpoint-9000_em':'em_without_retrieval'})
```

```python
joint_df = df_retrieval.merge(df_without_retrieval[["gen_answer_without_retrieval","f1_without_retrieval","rougeL_without_retrieval","em_without_retrieval"]], left_index=True, right_index=True)
joint_df.head()
```

```python
joint_df[["question", "gen_answer_news", "documents", "gen_answer_without_retrieval"]].to_csv("manual_news.csv")
```
